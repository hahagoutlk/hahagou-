```python
from pathlib import Path
import pandas as pd


INPUT_FILE = Path("/Users/tanglikun/Desktop/A股AI产业链召回总表_第二轮.xlsx")
OUTPUT_FILE = Path("/Users/tanglikun/Desktop/A股AI产业链召回总表_第三轮.xlsx")


THIRD_LAYER_KEYWORDS = {
    "AI基础层-算力基础设施补充": [
        "智算集群", "算力集群", "算力网络", "算网融合", "东数西算",
        "超算", "超级计算", "高性能计算", "HPC", "并行计算",
        "云原生", "私有云", "公有云", "混合云", "边缘云",
        "云资源", "云服务商", "云基础设施", "数据机房", "机房建设",
        "数据中心建设", "IDC机房", "服务器机柜", "机柜", "机房温控",
        "精密空调", "UPS电源", "备用电源", "配电柜", "母线槽"
    ],
    "AI基础层-半导体细分补充": [
        "算力芯片", "通用GPU", "加速卡", "AI加速卡", "推理卡",
        "计算卡", "异构计算", "RISC-V", "ARM架构", "SoC芯片",
        "神经网络处理器", "张量处理器", "边缘AI芯片",
        "车载芯片", "智能座舱芯片", "自动驾驶芯片",
        "存算一体", "Chiplet", "2.5D封装", "3D封装", "CoWoS",
        "TSV", "硅中介层", "晶圆级封装", "扇出型封装",
        "封测", "晶圆制造", "晶圆代工", "掩膜版", "电子特气",
        "光刻胶", "CMP", "靶材", "湿电子化学品"
    ],
    "AI基础层-光通信与网络补充": [
        "光收发模块", "光模块封装", "光引擎", "光电芯片",
        "激光器", "探测器", "硅光模块", "相干光模块",
        "高速率光模块", "400G", "800G", "1.6T光模块",
        "光纤连接器", "光纤光缆", "光网络", "数据中心交换机",
        "高速交换机", "白盒交换机", "网络可视化", "网络安全",
        "SDN", "网络控制器", "低时延网络"
    ],
    "AI基础层-PCB连接器材料补充": [
        "高频高速PCB", "服务器PCB", "AI服务器PCB", "高多层板",
        "高速覆铜板", "覆铜板", "CCL", "低介电材料",
        "高速材料", "电子布", "玻纤布", "铜箔", "高频材料",
        "高速背板", "背板连接器", "高速背板", "高速连接",
        "高速线束", "高速铜缆", "DAC", "ACC", "AEC",
        "液冷接头", "散热模组", "导热材料", "导热硅脂",
        "均热板", "热管", "散热器"
    ],
    "AI数据层-数据与软件基础": [
        "数据资产", "数据资源", "数据确权", "数据交易",
        "数据安全", "隐私计算", "联邦学习", "可信计算",
        "数据中台", "数据仓库", "数据湖", "湖仓一体",
        "向量检索", "向量化", "知识库", "企业知识库",
        "智能搜索", "语义搜索", "推荐算法", "搜索算法",
        "图数据库", "时序数据库", "分布式数据库",
        "低代码", "无代码", "智能运维", "AIOps",
        "DevOps", "软件定义", "数字孪生"
    ],
    "AI算法模型补充": [
        "视觉算法", "语音算法", "推荐算法", "预测算法",
        "优化算法", "路径规划", "SLAM", "目标检测",
        "目标识别", "图像识别", "图像处理", "视频分析",
        "行为识别", "人脸识别", "声纹识别", "语义理解",
        "文本生成", "智能问答", "对话系统", "认知智能",
        "决策智能", "智能决策", "专家系统"
    ],
    "AI应用-工业制造补充": [
        "工业智能", "智能产线", "柔性制造", "智能装备",
        "自动化产线", "工业软件", "MES", "PLM", "ERP",
        "SCADA", "DCS", "工业控制", "工控系统",
        "工业检测", "视觉检测", "质量检测", "缺陷识别",
        "良率提升", "预测维护", "设备健康管理",
        "智能调度", "智能排产", "无人产线"
    ],
    "AI应用-机器人零部件补充": [
        "谐波减速器", "RV减速器", "伺服电机", "伺服驱动",
        "控制器", "运动控制", "运动控制器", "编码器",
        "力传感器", "六维力传感器", "关节模组", "执行器",
        "线性执行器", "空心杯电机", "无框力矩电机",
        "丝杠", "滚珠丝杠", "行星滚柱丝杠", "灵巧手",
        "末端执行器", "机器人视觉", "机器人系统集成"
    ],
    "AI应用-汽车智能化补充": [
        "智能网联汽车", "车载智能", "车载摄像头",
        "车载传感器", "车载雷达", "车载激光雷达",
        "激光雷达", "毫米波雷达", "4D毫米波雷达",
        "高精度地图", "高精定位", "组合导航",
        "域控制器", "中央计算平台", "电子电气架构",
        "线控转向", "线控制动", "智能底盘",
        "座舱系统", "智能座舱系统", "HUD", "AR-HUD"
    ],
    "AI应用-医疗与生命科学": [
        "医学影像AI", "辅助诊断", "智能诊断", "影像诊断",
        "病理诊断", "AI辅助诊断", "智慧医院", "互联网医疗",
        "医疗机器人", "手术机器人", "康复机器人",
        "药物发现", "AI药物发现", "药物筛选", "蛋白质结构",
        "基因测序", "生物信息", "精准医疗"
    ],
    "AI应用-安防城市能源矿山": [
        "视频结构化", "视频智能分析", "智能监控", "智能安防",
        "城市大脑", "智慧园区", "智慧社区", "智慧停车",
        "智慧水务", "智慧环保", "智慧应急", "智慧消防",
        "智能电网", "电力巡检", "无人巡检", "智能巡检",
        "智慧矿山", "矿山智能化", "无人矿山", "智能采矿"
    ],
    "AI终端和交互补充": [
        "AI眼镜", "智能眼镜", "AR眼镜", "VR头显", "MR头显",
        "空间计算", "智能耳机", "智能音箱", "智能门锁",
        "智能家电", "智能座舱交互", "语音助手",
        "虚拟现实", "增强现实", "混合现实", "手势识别",
        "眼动追踪", "3D视觉", "深度相机"
    ],
}


WEAK_KEYWORDS = [
    "智能化", "数字化转型", "信息化建设", "平台建设",
    "系统平台", "管理平台", "监测平台", "调度平台",
    "自动控制", "自动识别", "数字经济"
]


def read_table(path):
    if path.suffix.lower() in [".xlsx", ".xls"]:
        return pd.read_excel(path, dtype=str)
    return pd.read_csv(path, dtype=str, encoding="utf-8-sig")


def find_hits_by_layer(text, keyword_dict):
    text = str(text)
    hit_keywords = []
    hit_layers = []

    for layer, keywords in keyword_dict.items():
        for kw in keywords:
            if kw in text:
                hit_keywords.append(kw)
                hit_layers.append(layer)

    return list(dict.fromkeys(hit_keywords)), list(dict.fromkeys(hit_layers))


def find_hits(text, keywords):
    text = str(text)
    return [kw for kw in keywords if kw in text]


def get_hit_text(text, hits, max_len=240):
    if not hits:
        return ""

    text = str(text).replace("\n", " ").replace("\r", " ")

    for kw in hits:
        pos = text.find(kw)
        if pos != -1:
            start = max(pos - 90, 0)
            end = min(pos + 150, len(text))
            return text[start:end][:max_len]

    return ""


def append_text(old, new):
    old = "" if pd.isna(old) else str(old)
    new = "" if pd.isna(new) else str(new)

    if not old:
        return new
    if not new:
        return old
    if new in old:
        return old
    return old + "；" + new


df = read_table(INPUT_FILE)
df = df.fillna("")

needed_cols = [
    "是否在AI产业链中",
    "召回来源",
    "命中关键词",
    "命中文本",
    "AI产业链环节",
    "确定性等级",
    "复核状态",
    "备注",
]

for col in needed_cols:
    if col not in df.columns:
        df[col] = ""

search_text = df.astype(str).agg(" ".join, axis=1)

step3_sources = []
step3_keywords = []
step3_texts = []
step3_layers = []

for i, text in enumerate(search_text):
    current_judgment = df.loc[i, "是否在AI产业链中"]

    strong_hits, layers = find_hits_by_layer(text, THIRD_LAYER_KEYWORDS)
    weak_hits = find_hits(text, WEAK_KEYWORDS)

    step3_keywords.append("、".join(strong_hits + weak_hits))
    step3_texts.append(get_hit_text(text, strong_hits + weak_hits))
    step3_layers.append("、".join(layers))

    if strong_hits:
        step3_sources.append("第三层扩展产业链关键词召回")

        df.loc[i, "是否在AI产业链中"] = "是"
        df.loc[i, "召回来源"] = append_text(df.loc[i, "召回来源"], "第三层扩展产业链关键词召回")
        df.loc[i, "命中关键词"] = append_text(df.loc[i, "命中关键词"], "、".join(strong_hits))
        df.loc[i, "命中文本"] = append_text(df.loc[i, "命中文本"], get_hit_text(text, strong_hits))
        df.loc[i, "AI产业链环节"] = append_text(df.loc[i, "AI产业链环节"], "、".join(layers))
        df.loc[i, "确定性等级"] = "中"
        df.loc[i, "复核状态"] = "待复核"
        df.loc[i, "备注"] = append_text(
            df.loc[i, "备注"],
            "第三层命中更细分AI产业链关键词，纳入候选池；后续需用主营业务、经营范围、年报或公告确认"
        )

    elif weak_hits:
        step3_sources.append("第三层弱相关关键词命中")

        if current_judgment != "是":
            df.loc[i, "是否在AI产业链中"] = "不能确定"
            df.loc[i, "召回来源"] = append_text(df.loc[i, "召回来源"], "第三层弱相关关键词命中")
            df.loc[i, "命中关键词"] = append_text(df.loc[i, "命中关键词"], "、".join(weak_hits))
            df.loc[i, "命中文本"] = append_text(df.loc[i, "命中文本"], get_hit_text(text, weak_hits))
            df.loc[i, "确定性等级"] = "低"
            df.loc[i, "复核状态"] = "进入下一轮"
            df.loc[i, "备注"] = append_text(
                df.loc[i, "备注"],
                "第三层仅命中弱相关词，不能直接确认，需要下一轮补充公司简介/主营业务/概念标签"
            )

    else:
        step3_sources.append("第三轮未召回")
        if current_judgment == "不能确定":
            df.loc[i, "复核状态"] = "进入下一轮"


df["第三层召回来源"] = step3_sources
df["第三层命中关键词"] = step3_keywords
df["第三层命中文本"] = step3_texts
df["第三层AI产业链环节"] = step3_layers

df.to_excel(OUTPUT_FILE, index=False)

print("第三轮 recall 完成！")
print(f"输出文件：{OUTPUT_FILE}")
print()
print("当前总判断结果：")
print(df["是否在AI产业链中"].value_counts())
print()
print("第三层召回情况：")
print(pd.Series(step3_sources).value_counts())
```
